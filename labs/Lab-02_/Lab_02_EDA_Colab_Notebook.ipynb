## 3Ô∏è‚É£ Shape (Distribution)

```

# Histogram of Tool Wear
plt.hist(df['VB'].dropna(), bins=20)
plt.title('Histogram of Tool Wear')
plt.xlabel('VB')
plt.ylabel('Frequency')
plt.show()
```

```

# KDE Plot
sns.kdeplot(df['VB'].dropna())
plt.title('KDE Plot of Tool Wear')
plt.show()
```
